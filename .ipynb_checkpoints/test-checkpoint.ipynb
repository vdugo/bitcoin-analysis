```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('../analysis/chatgpt_twitter_sentiment/tweets_pulled_100k_jan_9_2023')
```

```python
tweet_content = df['content']
tweet_content = tweet_content.dropna()
```

```python
tweet_content = list(tweet_content)
```

```python
# Create a string of all the tweet content
tweet_text = " ".join(tweet for tweet in tweet_content)
```

```python
# Create a word cloud object
wordcloud = WordCloud().generate(tweet_text)
```

```python
# Display the word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
